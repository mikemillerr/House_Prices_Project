```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_pickle('./preprocessed_df.pkl')
```

## Question: How can I get more money for my house?
## Customer: Seller
## Answer:
A very good option for somebody who wants to sell its house is to improve the grade of the house. __Add Example__


```python
f, ax = plt.subplots(figsize=(9, 5))
fig = sns.boxplot(x='grade', y="price", data=df)
fig.axis(ymin=0, ymax=5000000);
#plt.title('How does the Grade Influence the Selling Price')
plt.ylabel('House Price in USD');
plt.savefig('./presentation/grade_improvement.jpg', dpi=300)
```

## Question: How can I get more money for my house?

## Question: When is the best time to sell a house?


```python
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['price_per_sqft_liv'] = df['price'] / df['sqft_living']
fig, ax = plt.subplots(figsize=(15,10))
for i in range(1,13):
    ax.boxplot(df[df['month'] == i]['sqft_living'], positions=[i])
    
ax.plot(df.groupby('month')['price_per_sqft_liv'].median()*10)
plt.ylim([0, 4000])
plt.show()


#df[df['month'] == 1]['price']
```

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.plot(df.groupby('month')['price_per_sqft_liv'].median())
plt.ylabel('median(dollar/sqft_living)')
plt.xlabel('month')
plt.savefig('./presentation/month_to_buy.jpg', dpi=300)
```

## Question: Is there a best day to buy?

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.plot(df.groupby('day')['price_per_sqft_liv'].median())
plt.ylabel('median(dollar/sqft_living)')
plt.xlabel('day')
plt.savefig('./presentation/day_to_buy.jpg', dpi=300)
```
